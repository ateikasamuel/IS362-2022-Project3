```python
# Import the Sqlalchemy library, Sqlalchemy library's create engine method, and panda's library
import sqlalchemy as sqla
import pandas as pd
import sqlite3
from sqlalchemy import create_engine

# Create a query that pulls the Customers’ Last Name and First Name, and each customer’s purchased Track names and Album Titles. 
# Sort the information by Customer LastName then Customer FirstName. 

cust_sql = '''SELECT c.LastName, c.FirstName, i.InvoiceId, il.InvoiceId, t.Name, a.Title
         FROM Customer as c
         INNER JOIN Invoice as i
         ON c.CustomerId = i.CustomerId
         INNER JOIN InvoiceLine as il
         ON i.InvoiceId = il.InvoiceId
         INNER JOIN Track as t
         ON il.TrackId = t.TrackId
         INNER JOIN Album as a
         ON t.AlbumId = a.AlbumId
         ORDER BY c.LastName, c.FirstName'''

# Python function to connect to the Sqlite Database and python function to connect to the Sqlite database
# Return the SQLALCHEMY Engine Object
con = sqlite3.connect("D:/IS362_Project3/Chinook_Sqlite.sqlite")

# Read and display the Sqlite table 
sales = pd.read_sql_query(cust_sql , con=con)

sales
```
